```python
xtrain , xtest , ytrain , ytest = train_test_split(x,y , train_size=.8 , random_state=42 )
```

```python
xtrain = np.array(xtrain)
xtest = np.array(xtest)
```

```python
xtrain = xtrain.reshape(-1,5,6,1)
xtest  = xtest.reshape(-1,5,6,1)
print(xtrain.shape)
print(xtest.shape)
```

```python
cnn =models.Sequential([
    layers.Conv2D(30,(2,2) , activation="relu" , input_shape =(5,6,1)),
    layers.MaxPool2D(2,2),
    layers.Flatten(),
    layers.Dense(64,activation="relu"),
   layers.Dense(1, activation="sigmoid")

])
cnn.compile(optimizer="adam",
            loss = "binary_crossentropy" ,
            metrics=["accuracy"])
```

```python
cnn.fit(xtrain,ytrain , epochs=10)
```

```python
cnn.evaluate(xtest,ytest)
```
